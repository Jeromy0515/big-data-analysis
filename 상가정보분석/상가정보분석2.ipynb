```python
i = g.reset_index()

i = i.rename(columns = {'상호명': '상호수'})
i
```

```python
# 상권업종소분류명, 상호수로 barplot 그리기
b = i.sort_values(by='상호수', ascending=False)
plt.figure(figsize=(15, 4))
sns.barplot(data = b, ci=None, x='상권업종소분류명', y='상호수')
```

```python
# 상권업종소분류명이 학원-입시인 데이터만 가져와서 시군구명, 상호수를 가지고 barplot 그려보기
academy_sub = i[i['상권업종소분류명'] == '학원-입시']
plt.figure(figsize=(15, 4))
sns.barplot(data = academy_sub, x='시군구명', y='상호수')
```

```python
# catplot으로 그리기
# data = i
# 옵션추가 : 열이 시군구명으로 나오게 하기

sns.catplot(data = i, x = '상권업종소분류명', y = '상호수', kind = 'bar', col = '시군구명', col_wrap = 2, sharex = False, aspect = 2)
```
